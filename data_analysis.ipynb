```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


sns.set_style("whitegrid")

df = pd.read_csv('Iris.csv')

print("Iris Species")
print(df.head(15))

# Exploring the structure of the data set
print("Id Data Type: ", df["Id"].dtype)
print("Sepal Width Data Type: ",df["SepalWidthCm"].dtype)
# Checks the most common Iris Species
species_counts = df["Species"].value_counts()
most_common = species_counts.idxmax()
count = species_counts.max()
print("Most Common Iris Species:", most_common)

# Simulating Missing Values
df.loc[0, "SepalLengthCm"] = np.nan
df.loc[10, "Species"] = np.nan

# Check for missing values
print(df.isnull().values.any())

# Replacing Missing Values With Median
avg_sepal_length = df["SepalLengthCm"].mean()
df["SepalLengthCm"] = df["SepalLengthCm"].fillna(avg_sepal_length)

df_filled = df.fillna({
    'SepalLengthCm': df['SepalLengthCm'].mean(),  # Fill numeric with mean
    'Species': 'unknown'                        # Fill categorical with string
})

print("Original dataset: ", df)
print("After filling missing values: ")
df_filled.isnull().sum()


```

```python
# Basic Data Analysis
print(df.describe())
# Calegorical Groupings
mean_sepal_length = df.groupby('Species')['SepalLengthCm'].mean()
mean_sepal_width = df.groupby('Species')['SepalWidthCm'].mean()
mean_petal_width = df.groupby('Species')['PetalWidthCm'].mean()
mean_petal_length = df.groupby('Species')['PetalLengthCm'].mean()

print("Mean Sepal Length: ", mean_sepal_length)
print("Mean Sepal Width: ", mean_sepal_width)
print("Mean Petal Length: ", mean_petal_length)
print("Mean Petal Width: ", mean_petal_width)

# Patterns Based on Analysis
print("Conclusions Based on Data Analysis")
print("Iris-virginica has the greatest sepal length average.")
print("Iris-versicolor has the least sepal width average.")
print("Iris-virginica  has the greatest petal length avareage.")
print("Iris-setosa  has the least petal width avareage.")
```

```python
# Data Visualization
# Bar Graph
plt.bar(df["SepalLengthCm"], df["PetalLengthCm"], width=0.5)
plt.title("Sepal and Petal Length Correlation")
plt.xlabel("Sepal Length in Cm")
plt.ylabel("Petal Length in Cm")
plt.show()
# Histogram
plt.hist(df["PetalWidthCm"], bins=5, color='blue', edgecolor='black')
plt.title("Petal Length Distribution")
plt.xlabel("Petal Length in Cm")
plt.ylabel("Frequency")
plt.show()
# Scatter Plot
Species = df["Species"]
sns.scatterplot(data=df, x='SepalLengthCm', y='PetalLengthCm', hue=Species)
plt.title("Sepal Length and Petal Length Correlation")
plt.xlabel("Sepal Length in Cm")
plt.ylabel("Petal Length in Cm")
plt.show()
# Pie Chart
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=120)
plt.title("Iris Species Distribution")
plt.show()
```
